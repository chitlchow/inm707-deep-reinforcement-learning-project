# Result Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
training_data = pd.read_csv('training_history.csv')
training_data.drop(columns='Unnamed: 0', inplace=True)
```

```python
training_data.plot(x='Episodes', y=['Score'], kind='line')
```

```python
training_data.plot(x='Episodes', y='Epsilon')
```

```python
dqn_history = pd.read_csv('result-dataset/training_history_dqn.csv')
dqn_history.drop(columns="Unnamed: 0", inplace=True)
dqn_history
```

```python
dqn_history.plot(x='ep', y =['all_time_highest', 'mean_score_last25_ep'])
```

```python
dqn_history.plot(x='ep', y='ep_time')
```

```python
ddqn_history = pd.read_csv('result-dataset/training_history_ddqn.csv')
ddqn_history.drop(columns="Unnamed: 0", inplace=True)
ddqn_history
```

```python
plt.plot(ddqn_history.ep, ddqn_history.all_time_highest)
plt.plot(dqn_history.ep, dqn_history.all_time_highest)
plt.legend(['Double DQN', 'DQN'])
plt.title('Highest Scores Achieved During Training')
```

```python
plt.plot(ddqn_history.ep, ddqn_history.mean_score_last25_ep)
plt.plot(dqn_history.ep, dqn_history.mean_score_last25_ep)
plt.legend(['Double DQN', 'DQN'])
plt.title('Mean Score every 25 Episodes')
```

```python
plt.plot(ddqn_history.ep, ddqn_history.ep_time)
plt.plot(dqn_history.ep, dqn_history.ep_time)
plt.legend(['Double DQN', 'DQN'])
plt.title('Mean Episode Time every 25 episodes')
```

```python
ddqn_noisy_history = pd.read_csv('result-dataset/training_history_ddqn_prb.csv')
ddqn_noisy_history.drop(columns="Unnamed: 0", inplace=True)
ddqn_noisy_history
```

```python
plt.plot(ddqn_noisy_history.ep, ddqn_noisy_history.ep_time)
plt.plot(ddqn_history.ep, ddqn_history.ep_time)
plt.plot(dqn_history.ep, dqn_history.ep_time)
plt.legend(['Noisy Net', 'Double DQN', 'DQN'])
plt.title('Mean Episode Time every 25 episodes')
```

```python
plt.plot(ddqn_noisy_history.ep, ddqn_noisy_history.mean_score_last25_ep)
plt.plot(ddqn_history.ep, ddqn_history.mean_score_last25_ep)
plt.plot(dqn_history.ep, dqn_history.mean_score_last25_ep)
plt.legend(['Noisy Net', 'Double DQN', 'DQN'])
plt.title('Mean Score every 25 episodes')
```

```python
plt.plot(ddqn_noisy_history.ep, ddqn_noisy_history.all_time_highest)
plt.plot(ddqn_history.ep, ddqn_history.all_time_highest)
plt.plot(dqn_history.ep, dqn_history.all_time_highest)
plt.legend(['Noisy Net', 'Double DQN', 'DQN'])
plt.title('Highest Score')
```
